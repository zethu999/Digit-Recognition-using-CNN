```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Flatten, Conv2D, MaxPool2D
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator


from tensorflow.keras.datasets import mnist
```

```python
X_train, X_test, Y_train, Y_test  = train_test_split(x_final, y_final, test_size=0.1, random_state=5 )
```

```python
# Normalizaion
X_train = X_train.astype('float')/255
X_test =  X_test.astype('float')/255

Y_train = to_categorical(Y_train, num_classes=2)
Y_test = to_categorical(Y_test, num_classes=2)
```

```python
X_train, X_val, Y_train, Y_val  = train_test_split(X_train, Y_train, test_size=0.1, random_state=5 )
```

```python
import matplotlib.pyplot as plt            # for image vizualization

plt.figure(figsize=(5,1))
plt.imshow(X_val[0])
```

```python
print(X_train.shape)

print(X_val.shape)
```

```python
Y_val
```

```python


#### Model


model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same', 
                 activation ='relu', input_shape = (28,28,3)))
model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same', 
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))


model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))


model.add(Flatten())
model.add(Dense(256, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(2, activation = "softmax"))





```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam' , loss = "categorical_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(X_train,Y_train, batch_size=64, validation_data=(X_val,Y_val), epochs=10)
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(history.history['loss'], color='b', label="Training loss")
ax[0].plot(history.history['val_loss'], color='r', label="validation loss",axes =ax[0])
legend = ax[0].legend(loc='best', shadow=True)

ax[1].plot(history.history['accuracy'], color='b', label="Training accuracy")
ax[1].plot(history.history['val_accuracy'], color='r',label="Validation accuracy")
legend = ax[1].legend(loc='best', shadow=True)
```

```python
# predictions on testing set and valiadation with confusion matrix

Y_pred = model.predict(X_test)

Y_pred


```
